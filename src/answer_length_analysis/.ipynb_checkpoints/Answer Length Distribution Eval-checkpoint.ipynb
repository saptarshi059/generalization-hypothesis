```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def print_plots(dataset):
    if dataset == 'SQuAD':
        roberta_df = pd.read_pickle('../../zero_shot_predictions/squad/squadv1/csarron_roberta-base-squad-v1_squad_predictions.pkl')
        bidaf_df = pd.read_pickle('../../zero_shot_predictions/squad/squadv1/rc-bidaf_squad_predictions.pkl')
    else:
        roberta_df = pd.read_pickle('../../zero_shot_predictions/techqa/navteca_roberta-base-squad2_Saptarshi7_techqa-squad-style_predictions.pkl')
        bidaf_df = pd.read_pickle('../../zero_shot_predictions/techqa/rc-bidaf_Saptarshi7_techqa-squad-style_predictions.pkl')

    gold_chars = []
    roberta_pred_chars = []
    for row in roberta_df.itertuples():
        for ans in set(row.gold_answers): # We can get the gold characters from either dataframe since its the same
            gold_chars.append(len(ans))
        roberta_pred_chars.append(len(row.predictions))
        
    bidaf_pred_chars = []
    for row in bidaf_df.itertuples():
        bidaf_pred_chars.append(len(row.predictions))

    _, gold_bins = np.histogram(gold_chars)

    fig, axs = plt.subplots(1, 3, figsize=(15, 5))
    for ax, data, title in zip(axs, [gold_chars, roberta_pred_chars, bidaf_pred_chars], 
                               [f'Gold Answer lengths for {dataset}', 
                                f'Predicted Answer lengths of RoBERTa on {dataset}',
                                f'Predicted Answer lengths of BiDAF on {dataset}']):
        if data == gold_chars:
            col = 'violet'
        elif data == roberta_pred_chars:
            col = 'seagreen'
        else:
            col = 'Turquoise'

        _, _, bars = ax.hist(data, bins=gold_bins, edgecolor='black', alpha=0.7, color=col)        
        ax.bar_label(bars)
        ax.set_title(title)
        ax.set_xlabel('Answer Length')
        ax.set_ylabel('Number of Predictions')
        ax.grid(False)
        ax.set_xticks(gold_bins)

    plt.tight_layout()
    #plt.show()
    plt.savefig(f'{dataset}_len_distributions', transparent=True)
```

```python
print_plots('SQuAD')
print_plots('TechQA')
```
